### Clean the Data

Perhaps we’d like to know more another day, but the team is really interested in just the relationship between calories (`calories_day`) and weight. …and maybe gender.

Can you remove the other columns?

```python
# Remove ‘em.
df_food_columns = df_food[['calories_day', 'weight', 'Gender']]
```

What about `NaN`s? How many are there?

```python
# Count ‘em.
df_food_missing = df_food_columns.isnull().sum()
print(df_food_missing)

```

We gotta remove those `NaN`s―the entire row.

```python
# Drop ‘em.
df_food_removed = df_food_columns.dropna()
print(df_food_removed.info())
```

But what about the weird non-numeric values in the column obviously meant for numeric data?

Notice the data type of that column from when you got the types of all the columns?

If only we could convert the column to a numeric type and drop the rows with invalid values. 🤔

```python
# Fix that.
# Make an explicit copy to avoid SettingWithCopyWarning
df_food_columns = df_food[['calories_day', 'weight', 'Gender']].copy()

# Convert 'calories_day' to numeric and fill NaNs with the mean
df_food_columns['calories_day'] = pd.to_numeric(df_food_columns['calories_day'], errors='coerce')
df_food_columns['calories_day'] = df_food_columns['calories_day'].fillna(df_food_columns['calories_day'].mean())

# Convert 'weight' to numeric and fill NaNs with the mean
df_food_columns['weight'] = pd.to_numeric(df_food_columns['weight'], errors='coerce')
df_food_columns['weight'] = df_food_columns['weight'].fillna(df_food_columns['weight'].mean())

# Fill missing 'Gender' values with the mode
df_food_columns['Gender'] = df_food_columns['Gender'].fillna(df_food_columns['Gender'].mode()[0])
```

Now this data seems reasonably clean for our purposes! 😁

### Bonus #0

Apply the same steps as before to `df_salary`, but do not filter for any specific industry. Do perform the other data cleaning stuff, and get to a point where you can generate the minimum, mean, and maximum by state.

```python
# Start with the full dataset (df_manager_salary) - already has renamed columns
# Filter for USD currency only
# Start with the full dataset (df_manager_salary) - already has renamed columns
# Filter for USD currency only and make an explicit copy to avoid warnings
df_salary_usd = df_manager_salary[df_manager_salary['currency_format'] == 'USD'].copy()

# Clean up country names - standardize US variations
df_salary_usd['person_country_location'] = df_salary_usd['person_country_location'].replace({
    'United States of America': 'US',
    'United States': 'US',
    'U.S.': 'US',
    'USA': 'US',
    'U.S.A.': 'US'
})

# Use regex to catch additional variations
df_salary_usd['person_country_location'] = df_salary_usd['person_country_location'].str.replace(
    r'^(us|u\.s\.a\.?|united states( of america)?|usa)$', 'US', case=False, regex=True
)

# Convert salary to numeric
if df_salary_usd['salary_given'].dtype == 'O':
    df_salary_usd['salary_given'] = pd.to_numeric(df_salary_usd['salary_given'].str.replace(',', ''), errors='coerce')
else:
    df_salary_usd['salary_given'] = pd.to_numeric(df_salary_usd['salary_given'], errors='coerce')

# Fill NaN salaries with mean
df_salary_usd['salary_given'] = df_salary_usd['salary_given'].fillna(df_salary_usd['salary_given'].mean())

# Convert timestamp and filter for recent years
df_salary_usd['exact_time'] = pd.to_datetime(df_salary_usd['exact_time'], errors='coerce')
df_salary_usd_recent = df_salary_usd[df_salary_usd['exact_time'].dt.year.isin([2021, 2022, 2023])]

# Generate summary statistics by state, sorted by mean salary
df_salary_usd_recent.groupby('person_state_location')['salary_given'].agg(['min', 'mean', 'max']).sort_values('mean', ascending=False)
```

### Bonus #3

Show the quantiles instead of just minimum, mean, and maximum―say 0%, 5%, 25%, 50%, 75%, 95%, and 100%. Outliers may be deceiving.

Sort by whatever interests you―like say the *50th* percentile.

And throw in a count by state too. It would be interesting to know how many data points contribute to the figures for each state. (*Hint: your nice formatting from Bonus #1 might not work this time around.* 😜)

```python
def custom_quantiles(x):
    return pd.Series({
        'count': x.count(),
        '0%': x.quantile(0.00),
        '5%': x.quantile(0.05),
        '25%': x.quantile(0.25),
        '50%': x.quantile(0.50),
        '75%': x.quantile(0.75),
        '95%': x.quantile(0.95),
        '100%': x.quantile(1.00)
    })

# Generate quantile statistics for single states
quantile_summary = df_salary_single_states.groupby('person_state_location')['salary_given'].apply(custom_quantiles)

# Unstack to get a proper DataFrame structure, then sort by 50th percentile (median)
quantile_summary = quantile_summary.unstack().sort_values('50%', ascending=False)

# Display the table (note: styling won't work as nicely with this multi-level approach)
quantile_summary
```
